```python
import tensorflow as tf
from tensorflow import keras
```

```python
train_labels_path='/kaggle/input/cassava-leaf-disease-classification/train.csv'
labels=pd.read_csv(train_labels_path)
```

```python
labels.head()
```

```python
labels=labels.astype(str)
labels.dtypes
```

```python
train_data_gen=keras.preprocessing.image.ImageDataGenerator(validation_split=0.1)
train_images=train_data_gen.flow_from_dataframe(dataframe=labels,directory='/kaggle/input/cassava-leaf-disease-classification/train_images',
                                                x_col='image_id',y_col='label',class_mode='categorical',batch_size=128)
```

```python
x_batch,y_batch=next(train_images)
```

```python
import matplotlib.pyplot as plt
for i in range (0,5):
    image=x_batch[i]
    plt.imshow(image[:,:,0])
    plt.show()
    

```
